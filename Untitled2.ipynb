```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import norm

df = pd.read_csv('supermercado.csv')

print("Preço Normal")
print("Média:", df['Preco_Normal'].mean())
print("Mediana:", df['Preco_Normal'].median())
print("Moda:", df['Preco_Normal'].mode()[0])
print("Desvio Padrão:", df['Preco_Normal'].std())
print("Variância:", df['Preco_Normal'].var())

z_scores = np.abs(stats.zscore(df['Preco_Normal']))
outliers = df[z_scores > 3]
print("\nOutliers (Preço Normal):")
print(outliers)

plt.hist(df['Preco_Normal'], bins=30, density=True, alpha=0.6, color='g')

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, df['Preco_Normal'].mean(), df['Preco_Normal'].std())
plt.plot(x, p, 'k', linewidth=2)
plt.title('Distribuição Normal - Preço Normal')
plt.show()

preco = 100
media = df['Preco_Normal'].mean()
desvio = df['Preco_Normal'].std()
z = (preco - media) / desvio
probabilidade = norm.cdf(z)
print(f"Probabilidade de preço ser menor ou igual a {preco}: {probabilidade:.4f}")
```
